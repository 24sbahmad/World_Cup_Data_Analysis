## Every World Cup game the USA have played in

```sql

WITH Game_Result AS (
    SELECT
        home_team,
        away_team,
        home_score,
        away_score,
        date,
        CASE 
            WHEN home_score > away_score THEN 'Home Win'
            WHEN home_score < away_score THEN 'Away Win'
            ELSE 'Draw'
        END AS final_score
    FROM results
    WHERE tournament = 'FIFA World Cup'
)
SELECT * FROM Game_Result
WHERE home_team = 'United States'
    OR away_team = 'United States'
ORDER BY date DESC;
```

## Every USA Goalscorer in a USA Victory

```sql
WITH USA_WINS AS (
    SELECT
        date,
        home_team,
        away_team,
        CASE 
            WHEN home_score > away_score THEN 'Home Win'
            WHEN home_score < away_score THEN 'Away Win'
            ELSE 'Draw'
        END AS final_score
    FROM results
    WHERE tournament = 'FIFA World Cup'
        AND (home_team = 'United States' OR away_team = 'United States')
        AND (
            (home_score > away_score AND home_team = 'United States')
            OR 
            (away_score > home_score AND away_team = 'United States')
        )
)
SELECT
    USA_WINS.date AS date,
    USA_WINS.home_team AS home_team,
    USA_WINS.away_team AS away_team,
    goalscorers.team,
    goalscorers.scorer,
    goalscorers.minute,
    goalscorers.own_goal
FROM USA_WINS
JOIN goalscorers 
    ON USA_WINS.date = goalscorers.date
    AND USA_WINS.home_team = goalscorers.home_team
    AND USA_WINS.away_team = goalscorers.away_team
WHERE goalscorers.team = 'United States'
ORDER BY USA_WINS.date DESC, goalscorers.minute ASC;
```
