## Q1: Historically, how have the USA done in tournaments?

```sql

WITH Game_Result AS (
    SELECT
        home_team,
        away_team,
        home_score,
        away_score,
        date,
        CASE 
            WHEN home_score > away_score THEN 'Home Win'
            WHEN home_score < away_score THEN 'Away Win'
            ELSE 'Draw'
        END AS final_score
    FROM results
    WHERE tournament = 'FIFA World Cup'
)
SELECT * FROM Game_Result
WHERE home_team = 'United States'
    OR away_team = 'United States'
ORDER BY date DESC;
```

```sql
WITH Average_Goals AS (
    SELECT
        home_team,
        away_team,
        AVG(home_score) AS avg_home_goals,
        AVG(away_score) AS avg_away_goals,
        SUBSTRING(date, 1, 4) AS year
    FROM results
    WHERE tournament = 'FIFA World Cup'
        AND (home_team = 'United States' OR away_team = 'United States')
)
SELECT * FROM Average_Goals
WHERE home_team = 'United States'
    OR away_team = 'United States'
GROUP BY home_team, away_team, year
ORDER BY year DESC;
```
